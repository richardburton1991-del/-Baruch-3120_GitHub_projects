```python
from datetime import datetime
print(f'Run time: {datetime.now().strftime("%D %T")}')

import numpy as np
import pandas as pd
```

```python
def read_data_file(filename):
    try:
        df_raw = pd.read_csv(filename, sep=';')

        if df_raw.empty:
            raise Exception("File was read successfully, but it contains no data rows.")

        print(f"File loaded successfully: {filename}")
        return df_raw

    except Exception as e:
        print(f"Error reading file: {e}")
        return None
```

```python
def validate_dates(df, date_col):
    errors = []
    standardized_dates = []

    for idx, value in df[date_col].items():
        try:
            converted_date = pd.to_datetime(value)
            standardized_dates.append(converted_date)
        except:
            standardized_dates.append(pd.NaT)
            errors.append({
                "row_index": idx,
                "column": date_col,
                "invalid_value": value
            })

    df[date_col] = standardized_dates
    return df, errors
```

```python
def validate_numeric(df, num_col):
    errors = []

    for idx, value in df[num_col].items():
        try:
            val = float(value)

            if val < 0:
                df.at[idx, num_col] = np.nan
                errors.append({
                    "row_index": idx,
                    "column": num_col,
                    "invalid_value": value
                })
            else:
                df.at[idx, num_col] = val

        except:
            df.at[idx, num_col] = np.nan
            errors.append({
                "row_index": idx,
                "column": num_col,
                "invalid_value": value
            })

    return df, errors
```

```python
def clean_data(df):
    df = df.copy()

    df, date_errors = validate_dates(df, "session_date")
    df, num_errors = validate_numeric(df, "study_hours")

    all_issues = date_errors + num_errors

    bad_rows = df["session_date"].isna() | df["study_hours"].isna()
    bad_count = bad_rows.sum()

    df = df[~bad_rows]

    before = len(df)
    df = df.drop_duplicates(subset=["study_session_id"])
    dup_count = before - len(df)

    return df, all_issues, bad_count, dup_count
```

```python
EMPTY_FILE = 'https://raw.githubusercontent.com/vjavaly/Baruch-CIS-3120/main/data/study_logs_EMPTY.csv'
GOOD_FILE = 'https://raw.githubusercontent.com/vjavaly/Baruch-CIS-3120/main/data/study_logs.csv'
```

```python
df_empty = read_data_file(EMPTY_FILE)
```

```python
df_raw = read_data_file(GOOD_FILE)
```

```python
cleaned_df, issues, bad_rows, dups = clean_data(df_raw)

print("Cleaning Complete.")
print("Starting Rows:", len(df_raw))
print("Bad Rows:", bad_rows)
print("Duplicates Removed:", dups)
print("Final Rows:", len(cleaned_df))
```
